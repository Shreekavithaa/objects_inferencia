# Making few predictions with the loaded objectsv2 model

```python
def predict_actual(image):
    a = time.time()
    pred_labels = []
    pred = model.predict(image, batch_size=256)
    for i, prob in enumerate(pred[0]):
        if prob == 1:
            pred_labels.append((labels[i], prob))
    c = time.time()
    return pred_labels
```

```python
def pre_process_image(path):
    #pre-processing the image
    image = np.array(skimage.io.imread(path))
    image = cv2.resize(image, (300, 300))
    image = np.array(image)
    image = image[:, :, ::-1]
    image = np.expand_dims(image, axis=0)
    shape = image.shape
    return image
```

```python
#example_input
predict_actual(pre_process_image("https://pbs.twimg.com/media/Es-xFCWUUAY15Ew.jpg"))
```

```python
# Load saved model
# use filenames according to the approach used

SAVED_MODEL_DIR = './ws_objects/objects/'
predictor = tf.contrib.predictor.from_saved_model(SAVED_MODEL_DIR)
```

```python
# Load compiled model
# objects_neuron_compiled - this has an on neuron ratio of 0.0

COMPILED_MODEL_DIR = './ws_objects/objects_neuron_1/'
predictor_inf = tf.contrib.predictor.from_saved_model(COMPILED_MODEL_DIR)
```

```python
from keras import layers
from tensorflow.keras.applications import resnet50

input_image = pre_process_image("https://pbs.twimg.com/media/Es-xFCWUUAY15Ew.jpg")
input_image = resnet50.preprocess_input(input_image)

test_image = pre_process_image("https://pbs.twimg.com/media/Et9KZ-RXcAEbghU?format=jpg&name=medium")
test_image = resnet50.preprocess_input(test_image)
```

```python
def predict(model_feed_dict, predictor, output_layer_name):
    a = time.time()
    pred_labels = []
    pred = predictor(model_feed_dict);
    for i, prob in enumerate(pred[output_layer_name][0]):
        if prob == 1.0:
            pred_labels.append(labels[i])
    c = time.time()
    return pred_labels
```

```python
# Run inference for saved model

model_feed_dict={'input': input_image}
predict(model_feed_dict, predictor_inf, "output")
```

```python
# Run inference for compiled model

model_feed_dict={'input': test_image}
predict(model_feed_dict, predictor_inf, "output")
```
